```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

##1.2 Arrays
Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
def sum_arr(arr):
  """
  This function takes an array (list) as input and returns the sum of its elements.
  Time Complexity: O(N), where N is the number of elements in the array.
  The function iterates through each element of the array once.
  """
  total = 0
  for element in arr:
    total += element
  return total
```

```python
def max_min(arr):
  """
  This function takes an array (list) as input and returns the maximum element.
  Time Complexity: O(N), where N is the number of elements in the array.
  The function iterates through each element of the array once to find the maximum.
  """
  if not arr:
    return None # Handle empty array case
  maximum = arr[0]
  for element in arr:
    if element > maximum:
      maximum = element
  return maximum
```

```python
arr = [12, 1234, 45, 67, 1]
print(f"Sum of elements in {arr}: {sum_arr(arr)}")
```

```python
def Counting_EvenOdd(arr):
  """
  This function takes an array (list) as input and counts the number of even and odd elements.
  Time Complexity: O(N), where N is the number of elements in the array.
  The function iterates through each element of the array once.
  """
  even_count = 0
  odd_count = 0
  for element in arr:
    if element % 2 == 0:
      even_count += 1
    else:
      odd_count += 1
  return even_count, odd_count
```

```python
def linear_Search(arr, target):
  """
  This function performs a linear search on a 1D array to find a target element.
  Time Complexity: O(N), where N is the number of elements in the array.
  In the worst case, the function might have to iterate through all elements.
  """
  for i in range(len(arr)):
    if arr[i] == target:
      return i  # Return the index if target is found
  return -1 # Return -1 if target is not found
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
```

```python
arr_search = [1, 3, 5, 7, 9]
target = 5
index = linear_Search(arr_search, target)

if index != -1:
  print(f"Target {target} found at index {index} in {arr_search}")
else:
  print(f"Target {target} not found in {arr_search}")
```

```python
def Binary_Search(arr, target):
  """
  This function performs a binary search on a sorted 1D array to find a target element.
  Time Complexity: O(log N), where N is the number of elements in the array.
  Assumes the input array 'arr' is already sorted.
  """
  low, high = 0, len(arr) - 1
  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == target:
      return mid  # Return the index if target is found
    elif arr[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
  return -1 # Return -1 if target is not found
```

```python
arr_sorted = [1, 3, 5, 7, 9, 14, 15, 16, 45]
target = 15
index = Binary_Search(arr_sorted, target)

if index != -1:
  print(f"Target {target} found at index {index} in {arr_sorted}")
else:
  print(f"Target {target} not found in {arr_sorted}")
```
